```python
df2_task = pd.read_csv('task5.csv', index_col=[0, 1, 2], header=[0, 1]) # не редактируйте ячейку
```

```python
df2_result = df2_task.swaplevel().swaplevel(1,0)
```

```python
df2_result
```

```python
# сортировка:

# Отсортируйте так чтобы сначала шли бесплатные парковки в алфавитном порядке по округу, а внутри 
# округа в алфавитном порядке по району*.

# *Некоторые названия районов записаны наоборот: вместо район Нижегородский записано Нижегородский район.
#     Из-за этого сортировка по алфавиту будет идти не совсем так, как вы ожидали. Оставим результат 
#     сортировки по алфавиту как есть (переносить слово район для корректной сортировки не будем).



# можно в sort_index указывать кортеж (level=(0,1,2))



df2_result.sort_index(level = 0)
```

Сохраните данные из этого уровня в отдельную колонку под названием `type`. Из целого датафрейма вырезаем только данные о рейтинге, сортируем записи по среднему рейтингу в порядке убывания и загружаем на степик ТОП-10 записей.

```python
df3_task = pd.read_csv('task5.csv', index_col=[0, 1, 2], header=[0, 1]) # не редактируйте ячейку
```

```python
df3_task
```

```python
df3_task.index.names
```

```python
# оставляем только rating, сортируем по mean на убывание

df3_task['rating'].sort_values(by = 'mean', ascending = False)
```

```python
# плюс сбрасываем индексы

df3_task['rating'].sort_values(by = 'mean', ascending = False).reset_index()
```

```python
# присваиваем результат переменной r

r = df3_task['rating'].sort_values(by = 'mean', ascending = False).reset_index()
```

```python
# устанавливаем индексы:

r.set_index(['level_0', 'level_1'])
```

```python
rr = r.set_index(['level_0', 'level_1'])
```

```python
# задаем пустые индексы

rr.index.names = [None, None]
```

```python
# именуем колонки

rr.columns = ['type', 'min', 'max', 'mean']
```

```python
df3_result = rr.head(10)
```
