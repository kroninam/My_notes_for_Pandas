# Начинаем с Series

```python
s1 = pd.Series([0, None, 2, 3, 4, None, 6, 7, 8, None],
               index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'i', 'h', 'k'])

s2 = pd.Series([None, 1, 2, None, 4, 5, 6, None, 8, None],
               index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'i', 'h', 'k'])
```

```python
s1
```

```python
s2
```

```python
s1.fillna(s2)
```

```python
np.where(pd.isnull(s1), s2, s1)
```

```python
s1.combine_first(s2)
```

```python
# если в s2 присутствует больше значений и инедексов чем в s1: 
# функция where будет давать ошибку,
# а combine_first отработает: - добавит в результат всё.
    
s1 = pd.Series([0, None, 2, 3, 4, None, 6, 7, 8, None],
               index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'i', 'h', 'k'])

s2 = pd.Series([None, 1, 2, None, 4, 5, 6, None, 8, None, 100],
               index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'i', 'h', 'k', 'w'])
```

```python
s1.combine_first(s2)
```

# Функции pivot и melt - "Широкая" и "Длиннная" таблица

```python
data = {'Год': [2018, 2019, 2020, 2021],

                    'Земельный налог': [5000, 6000, 4000, 2000],

                     'Налог на доб. стоимость': [20000,  21000, 22000, 23000],

                     'Налог на доход ФЛ': [180000,  190000,  200000, 210000],

                     'Налог на имущество': [54000, 54500,  55000,  55500],

                     'Налог на прибыль': [490000,  520000,  550000,  580000],

                     'Транспортный налог': [25000,  20000,  28000, 20000]}

df = pd.DataFrame(data)
```

```python
# Это широкая таблица
df
```

```python
# Переведем широкую таблицу в длинную:


# год закрепляем, все остальное сползает в длинну.

pd.melt(df, 'Год')
```

```python
# теперь имеем длинную таблицу и хотим перевести ее обратно в широкую:

df1 = pd.melt(df, 'Год')
```

```python
df1.pivot(values = 'value', columns = 'variable', index = 'Год')
```

```python
df1 = df1.pivot(values = 'value', columns = 'variable', index = 'Год')
```

```python
df1['Земельный налог']
```

```python
# еще можно так. Задаем мульти индекс:
df2 = df1.set_index(['Год', 'variable'])
df1.set_index(['Год', 'variable'])
```

```python
df2.unstack()
```
