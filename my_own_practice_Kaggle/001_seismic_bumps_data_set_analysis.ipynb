```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

# Ð—Ð°Ð³Ñ€ÑƒÐ·Ð¸Ð¼ Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð¸ ÑÐ¾Ð·Ð´Ð°Ð´Ð¸Ð¼ Ð´Ð°Ñ‚Ð°Ñ„Ñ€ÐµÐ¹Ð¼
# data = pd.read_csv('data.csv')
# df = pd.DataFrame(data, columns=['x1', 'x2', 'y'])

data = pd.read_csv('seismic-bumps.csv')
df = pd.DataFrame(data, columns=['genergy', 'gpuls', 'energy'])


# Ð”Ð¾Ð±Ð°Ð²Ð¸Ð¼ ÐºÐ¾Ð½ÑÑ‚Ð°Ð½Ñ‚Ñƒ Ð´Ð»Ñ Ð¼Ð¾Ð´ÐµÐ»Ð¸
df = sm.add_constant(df)

# Ð¡Ð¾Ð·Ð´Ð°Ð´Ð¸Ð¼ Ð¾Ð±ÑŠÐµÐºÑ‚ Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð¹ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸
model = sm.OLS(df['gpuls'], df[['const', 'genergy', 'energy']])

# ÐžÐ±ÑƒÑ‡Ð¸Ð¼ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð¸ Ð¿Ð¾Ð»ÑƒÑ‡Ð¸Ð¼ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ñ‹
results = model.fit()

# Ð’Ñ‹Ð²ÐµÐ´ÐµÐ¼ ÑÐ²Ð¾Ð´Ð½ÑƒÑŽ Ð¸Ð½Ñ„Ð¾Ñ€Ð¼Ð°Ñ†Ð¸ÑŽ Ð¾ Ð¼Ð¾Ð´ÐµÐ»Ð¸
print(results.summary())

# ÐŸÐ¾ÑÑ‚Ñ€Ð¾Ð¸Ð¼ Ð³Ñ€Ð°Ñ„Ð¸Ðº
fig, ax = plt.subplots()
ax.scatter(df['genergy'], df['gpuls'], label='genergy')
ax.scatter(df['energy'], df['gpuls'], label='energy')
ax.plot(df['genergy'], results.fittedvalues, color='red', label='Fitted Line')
ax.legend()

# Ð”Ð¾Ð±Ð°Ð²Ð¸Ð¼ ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚ ÐºÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ð¸ Ð½Ð° Ð³Ñ€Ð°Ñ„Ð¸Ðº
corr = np.corrcoef(df['genergy'], df['energy'])[0, 1]
ax.text(0.05, 0.95, f'Correlation = {corr:.2f}', transform=ax.transAxes)

plt.show()
```
