```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Создаем данные
# x = np.array([1, 2, 3, 4, 5])
# y = np.array([2.5, 3.7, 4.1, 4.8, 6.0])

x = df.genergy
y = df.gpuls

# Вычисляем коэффициенты линейной регрессии
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Создаем scatter plot
fig, ax = plt.subplots()
ax.scatter(x, y)

# Добавляем линию тренда
ax.plot(x, slope*x + intercept, color='red')

# Добавляем уравнение зависимости и коэффициент корреляции
ax.text(0.1, 0.9, f'y = {slope:.2f}x + {intercept:.2f}\nR = {r_value:.2f}', transform=ax.transAxes)


# Подписи осей
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter plot of X vs Y & trend')



# Отображаем график
plt.show()
```

```python
Проверка на нормальность распределения

Здесь мы используем функцию norm из библиотеки scipy.stats, 
чтобы сгенерировать нормально распределенные данные. 
Затем мы строим гистограмму распределения с помощью 
метода hist и проверяем нормальность распределения с 
помощью функции normaltest. Если значение p-value меньше 
0.05, то мы считаем, что распределение не является нормальным, 
и выводим соответствующее сообщение на график с помощью метода text.

Обратите внимание, что в примере мы используем 30 бинов для гистограммы, но это значение может быть изменено в 
зависимости от количества данных и требуемой точности гистограммы. Также стоит заметить, что тест на нормальность 
не является абсолютной проверкой на нормальность, поэтому, если вам нужна более точная проверка, можете использовать 
другие тесты или методы.

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, normaltest

# Создаем нормально распределенные данные
data = norm.rvs(size=1000)

# Строим гистограмму распределения
fig, ax = plt.subplots()
ax.hist(data, bins=30)

# Проверяем нормальность распределения
stat, p = normaltest(data)
if p < 0.05:
    ax.text(0.05, 0.95, f'p = {p:.2f}\nNon-normal', transform=ax.transAxes, va='top')
else:
    ax.text(0.05, 0.95, f'p = {p:.2f}\nNormal', transform=ax.transAxes, va='top')

# Отображаем график
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

# Загрузим данные и создадим датафрейм
# data = pd.read_csv('data.csv')
# df = pd.DataFrame(data, columns=['x1', 'x2', 'y'])

data = pd.read_csv('seismic-bumps.csv')
df = pd.DataFrame(data, columns=['genergy', 'gpuls', 'energy'])


# Добавим константу для модели
df = sm.add_constant(df)

# Создадим объект модели линейной регрессии
model = sm.OLS(df['gpuls'], df[['const', 'genergy', 'energy']])

# Обучим модель и получим результаты
results = model.fit()

# Выведем сводную информацию о модели
print(results.summary())

# Построим график
fig, ax = plt.subplots()
ax.scatter(df['genergy'], df['gpuls'], label='genergy')
ax.scatter(df['energy'], df['gpuls'], label='energy')
ax.plot(df['genergy'], results.fittedvalues, color='red', label='Fitted Line')
ax.legend()

# Добавим коэффициент корреляции на график
corr = np.corrcoef(df['genergy'], df['energy'])[0, 1]
ax.text(0.05, 0.95, f'Correlation = {corr:.2f}', transform=ax.transAxes)

plt.show()
```
